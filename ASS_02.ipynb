```python
# ------------------------------ PRACTICAL: FEEDFORWARD NEURAL NETWORK ------------------------------

# 1️⃣ Import necessary libraries
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np

# 2️⃣ Load and normalize the data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0

# 3️⃣ Build the Feedforward Neural Network
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))          # Convert 28x28 image → 784 vector
model.add(Dense(128, activation='relu'))          # Hidden layer with 128 neurons
model.add(Dense(10, activation='softmax'))        # Output layer for 10 classes (0–9)

# 4️⃣ Compile the model
model.compile(optimizer='sgd',                    # SGD = Stochastic Gradient Descent
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5️⃣ Train the model
history = model.fit(x_train, y_train,
                    validation_data=(x_test, y_test),
                    epochs=10)                    # Train for 10 epochs

# 6️⃣ Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy: %.2f%%" % (test_acc * 100))
print("Test Loss:", test_loss)



# 8️⃣ Predict the output of a test image
plt.imshow(x_test[0], cmap='plasma')
plt.title("Test Image")
plt.show()

prediction = model.predict(x_test)
print("The handwritten number in the image is:", np.argmax(prediction[0]))
```
