```python
# -------------------- STEP 1: IMPORT LIBRARIES --------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import regularizers
from google.colab import files
import io
# -------------------- STEP 2: UPLOAD DATA --------------------
print("Upload your dataset CSV file")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
data = pd.read_csv(io.BytesIO(uploaded[filename]))
display(data.head())
# -------------------- STEP 3: CLEAN AND SCALE DATA --------------------
data_numeric = data.select_dtypes(include=[np.number])  # Only numeric
data_numeric = data_numeric.fillna(data_numeric.median())  # Handle missing
scaler = MinMaxScaler()  # Scale between 0–1
data_scaled = scaler.fit_transform(data_numeric)
# -------------------- STEP 4: BUILD ENCODER-DECODER --------------------
input_dim = data_scaled.shape[1]         # Number of columns
encoding_dim = max(2, input_dim // 2)    # Compressed size

input_layer = Input(shape=(input_dim,))  # Input layer
encoder = Dense(encoding_dim, activation='relu',
                activity_regularizer=regularizers.l1(1e-5))(input_layer)
decoder = Dense(input_dim, activation='sigmoid')(encoder)
autoencoder = Model(inputs=input_layer, outputs=decoder)
# -------------------- STEP 5: COMPILE MODEL --------------------
autoencoder.compile(optimizer='adam', loss='mse', metrics=['mae'])
# -------------------- STEP 6: TRAIN MODEL --------------------
history = autoencoder.fit(data_scaled, data_scaled,
                          epochs=50,
                          batch_size=32,
                          shuffle=True,
                          validation_split=0.1,
                          verbose=1)
# -------------------- STEP 7: RECONSTRUCT DATA --------------------
reconstructed = autoencoder.predict(data_scaled)
# -------------------- STEP 8: CALCULATE RECONSTRUCTION ERROR --------------------
mse = np.mean(np.power(data_scaled - reconstructed, 2), axis=1)
# -------------------- STEP 9: SET THRESHOLD --------------------
threshold = np.percentile(mse, 95)
print("Reconstruction error threshold:", threshold)
import matplotlib.pyplot as plt

plt.hist(mse, bins=20, color='lightblue')
plt.axvline(threshold, color='red')
plt.xlabel('Reconstruction Error')
plt.ylabel('Count')
plt.title('Reconstruction Error Distribution')
plt.legend()
plt.show()

# -------------------- STEP 11: DETECT ANOMALIES --------------------
anomalies = mse > threshold
print("Number of anomalies:", np.sum(anomalies))
print("Anomaly indices:", np.where(anomalies)[0][:10])
display(data.iloc[np.where(anomalies)[0][:10]])
```
